```python
# Food Delivery Data Analysis

This notebook analyzes food delivery data to answer analytical and numerical
questions.
```

```python
# Data Loading


import pandas as pd
import json
import sqlite3

#We Load orders
orders = pd.read_csv("orders.csv")

#We Load users
with open("users.json") as f:
    users = pd.DataFrame(json.load(f))

#We Load restaurants from SQL
conn = sqlite3.connect(":memory:")
with open("restaurants.sql") as f:
    conn.executescript(f.read())

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)

orders.head(), users.head(), restaurants.head()

```

```python
#Data Merging

df = orders.merge(users, on="user_id").merge(restaurants, on="restaurant_id")
df.head()
```
